```python
def graph_to_hierarchical(G, levels=1):
    nodes_by_count = sorted([(d['count'], x) for x, d in G.nodes_iter(data=True)], reverse=True)
    if levels == 0:
        children = []
        for x, d in G.nodes(data=True):
            node_data = d
            node_data['name'] = x
            children.append(node_data)
    else:
        partition = community.best_partition(G)
        no_of_communities = max(partition.values()) + 1
        Gcoms = [G.subgraph([x for x in G.nodes_iter() if partition[x] == i])
                 for i in range(no_of_communities)]
        children = [graph_to_hierarchical(Gcom, levels=levels-1)
                    for Gcom in Gcoms]
    return {'name': nodes_by_count[0][1],  # or maybe more?
            'children': children}
```

```python
hierarchy = graph_to_hierarchical(G, levels=2)
```

```python
json.dump(hierarchy, open("../hierarchic/so_2048.json", "w"), indent=2)
```
